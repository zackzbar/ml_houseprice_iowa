```python
import numpy as np
import pandas as pd
```

## Make Full Dummy Data File

```python
data = pd.read_csv('../processed_data/train_full.csv')
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

```python
# full list of dummies
dummy_list = ['MSSubClass','MSZoning','Street','Alley','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','OverallQual','OverallCond','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','CentralAir','Electrical','KitchenQual','Functional','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','PoolQC','Fence','MiscFeature','SaleType','SaleCondition']
len(dummy_list)
```

```python
# make function that will make dummy dataframe
def make_dummy_df(df, cols) :
    for x in cols:
        dums = pd.get_dummies(df[x], drop_first=True, prefix=x)
        df = pd.concat((df, dums), axis=1)
        df.drop([x], axis=1, inplace=True)
    return df
```

```python
# call it
data_dum = make_dummy_df(data, dummy_list)
```

```python
# check
print(data_dum.shape)
data_dum.head()
```

```python
# Write to CSV
data_dum.to_csv('train_dummy.csv')
```

## Make MLR Data File

```python
# Now for mlr data

mlr_list = ['Id',
 'OverallQual',
 'CentralAir',
 'OverallCond',
 'KitchenQual',
 'GarageCars',
 'Functional', 
 'Fireplaces',
 'BsmtExposure', 
 'GarageFinish',
 'GarageQual',
 'PavedDrive',
 'BsmtQual',
 'HeatingQC',
 'ExterQual',
 'FireplaceQu',
 'BsmtFinType1',
 'BsmtFinType2',
 'TotRmsAbvGrd',
 'BsmtFinSF1', 
 'MasVnrArea',
 'BsmtFinSF2',
 'KitchenAbvGr',
 'Outdoor_Rec',
 'Total_Fullbath',
 'Total_Halfbath',
 'Total_House_Area',
 'FinBsmt',
 'SalePrice',
 'logSalePrice']

mlr_dum_list = ['CentralAir',
 'PavedDrive']

print(len(mlr_list)-1)
print(len(set(mlr_list))-1)
print(len(mlr_list)-1-len(mlr_dum_list))

from collections import Counter
[k for k,v in Counter(mlr_list).items() if v>1]
```

```python
# Feature engineering
mlr_data = pd.read_csv('../processed_data/train_ordinal.csv')
mlr_data['Outdoor_Rec'] = mlr_data['WoodDeckSF'] + mlr_data['OpenPorchSF'] + mlr_data['EnclosedPorch'] + mlr_data['3SsnPorch'] + mlr_data['ScreenPorch']
mlr_data['Total_Fullbath'] = mlr_data['FullBath'] + mlr_data['BsmtFullBath']
mlr_data['Total_Halfbath'] = mlr_data['HalfBath'] + mlr_data['BsmtHalfBath']
mlr_data['Total_House_Area'] = mlr_data['GrLivArea'] + mlr_data['TotalBsmtSF']
mlr_data['FinBsmt'] = 1-(mlr_data['BsmtUnfSF'] / mlr_data['TotalBsmtSF'])
mlr_data['logSalePrice'] = mlr_data.SalePrice.map(lambda x: np.math.log(x))
mlr_data['FinBsmt'].fillna(0, inplace=True)
```

```python
# call dummy function

mlr_data = mlr_data[mlr_list].copy()
display(mlr_data)
mlr_dum = make_dummy_df(mlr_data, mlr_dum_list)
mlr_dum
```

```python
# Write to CSV
mlr_dum.to_csv('../processed_data/train_mlr2.csv')
```
